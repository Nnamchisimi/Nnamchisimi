```python
import pandas as pd

# Paths for the first set of files
reference_file_path1 = '/content/latestStarOrderplus_-_Order_items_20240730094341 (2).xlsx'
target_file_path1 = '/content/StarOrderplus_-_Order_items_20250120081608.xlsx'
output_file_path1 = '/content/updated_Order_items.xlsx'

# Paths for the second set of files
reference_file_path2 = '/content/latestMercedes Yedek Parça Sipariş 2024.xlsx'
target_file_path2 = '/content/Mercedes Yedek Parça Sipariş 2024.xlsx'
output_file_path2 = '/content/updated_Mercedes_Yedek_Parça_Sipariş_2024.xlsx'



# Function to update the headings of a target file
def update_headings(reference_file_path, target_file_path, output_file_path):
    # Read the reference file to get the standard headings
    reference_df = pd.read_excel(reference_file_path)
    standard_headings = reference_df.columns.tolist()

    # Read the target file
    target_df = pd.read_excel(target_file_path)

    # Replace the column headings with the standard headings
    target_df.columns = standard_headings

    # Save the modified file to the specified output file path
    target_df.to_excel(output_file_path, index=False)

# Update the headings of the first set of files and save the result in a new file
update_headings(reference_file_path1, target_file_path1, output_file_path1)
print(f"Headings updated successfully for the first file and saved to {output_file_path1}.")

# Update the headings of the second set of files and save the result in a new file
update_headings(reference_file_path2, target_file_path2, output_file_path2)
print(f"Headings updated successfully for the second file and saved to {output_file_path2}.")
```

```python
import pandas as pd

# Load the file
file_path = "/content/orders-overview-detail-export (2).xlsx"
df = pd.read_excel(file_path)

# Step 1: Clean 'Group' column by stripping all leading/trailing spaces
df['Group'] = df['Group'].str.strip()

# Step 2: Create 'Base_Group' by removing the last character ('0' or '1') for comparison
df['Base_Group'] = df['Group'].str[:-1]

# Step 3: Identify rows where 'Group' ends with '1' and remove those rows
# Create a mask for groups ending with '1' and their corresponding 'Base_Group'
rows_to_remove = df[df['Group'].str.endswith('1')]

# Step 4: Remove the rows with 'Group' ending with '1'
df = df[~df['Group'].isin(rows_to_remove['Group'])]

# Step 5: Drop the 'Base_Group' column as it is no longer needed
df.drop(columns=['Base_Group'], inplace=True)

# Step 6: Date columns to convert to the required format
date_columns = ["Ord. date", "Requested d.", "Conf. DD", "Expected DD", "DN date", "Inv. date"]

# Step 7: Convert the date columns to the required format
def convert_date_format(df, columns):
    for column in columns:
        if column in df.columns:
            try:
                # Convert to datetime and reformat
                df[column] = pd.to_datetime(df[column], errors='coerce', format='%m/%d/%Y %I:%M:%S %p')
                df[column] = df[column].dt.strftime('%d/%m/%Y %I:%M %p')
            except Exception as e:
                print(f"Error processing column '{column}': {e}")
    return df

# Apply date conversion to the DataFrame
df = convert_date_format(df, date_columns)

# Step 8: Save the cleaned DataFrame to a new file
output_file_path = "/content/cleaned_orders-overview-detail-export.xlsx"
df.to_excel(output_file_path, index=False)

print(f"Cleaned file saved to {output_file_path}")
```

```python
import pandas as pd

# Function to update the column headings of the target file based on the reference file
def update_headings(reference_file_path, target_file_path, output_file_path):
    # Read the reference file to get the standard column headings
    reference_df = pd.read_excel(reference_file_path)

    # Read the target file that needs column headings updated
    target_df = pd.read_excel(target_file_path)

    # Get the list of standard column headings from the reference file
    standard_headings = reference_df.columns.tolist()

    # Update the column headings in the target file to match the standard headings
    target_df.columns = standard_headings

    # Save the modified file with updated headings to the output file path
    target_df.to_excel(output_file_path, index=False)

    # Print a message indicating that the file has been updated
    print(f"Headings updated successfully for the file: {target_file_path} and saved to {output_file_path}.")

# Define the path for the first input file
file1 = "/content/orders-overview-detail-export (6).xlsx"
df1 = pd.read_excel(file1)

# Define the column mapping for renaming the columns in df1
column_mapping = {
    "gATP": "gATP",
    "Code": "Code",
    "Customer no.": "Cust. No.",
    "Order type": "Ord. type",
    "Distribution channel": "Dist. ch.",
    "Customer order no.": "Cust. order no.",
    "CRM order no.": "CRM order no.",
    "Original item no.": "Orig. item no.",
    "Customer info": "Customer info",
    "Code list": "Code list",
    "Part no ordered": "PN ordered",
    "Part description (ordered)": "Designation of part no. ordered",
    "Part no. confirmed": "PN confirmed",
    "Description of part no. confirmed": "Designation of part no. confirmed",
    "Item status": "Items status",
    "Quantity ordered": "Q ordered",
    "Quantity confirmed": "Q confirm.",
    "Quantity goods issue": "Q in comm.",
    "Quantity expected": "Q expect.",
    "Quantity cancelled": "Q canceled",
    "Creation Date": "Ord. date",
    "Requested date": "Requested d.",
    "Confirmed despatch date": "Conf. DD",
    "Expected despatch date": "Expected DD",
    "Delivery site": "Deliv. site",
    "Alternative goods recipient": "Altern. goods recipient",
    "Requirement rule": "Req. rule",
    "Replacement rule": "Repl. rule",
    "Chassis no.": "Chas. no.",
    "Additional entry": "Additional entry",
    "Item remarks": "Item remarks",
    "Delivery note no": "DN no.",
    "Delivery note item": "DN item",
    "Delivery note date": "DN date",
    "Invoice no.": "Inv. no.",
    "Invoice Item": "Inv. item",
    "Invoice date": "Inv. date",
    "Item net value": "Value",
    "Discount group": "DG",
    "Campaign no.": "Camp. no.",
    "Volume": "Vol.",
    "Weight": "Wght",
    "Duty rate no.": "Duty rate no.",
    "Country of manufacture": "Region of manuf.",
    "Hazard class": "Hazard cl.",
    "Marketing code": "M code",
    "Shelf life": "Shelf life",
    "Return excluded": "Return excluded",
    "DRT Identifier": "DRT ID",
    "Costing code": "Costing code",
    "Hub indicator": "Hub ind.",
    "Closing order no.": "CON",
    "Situation report": "Situation report"
}

# Rename the columns in df1 based on the defined column mapping
df1.rename(columns=column_mapping, inplace=True)

# Clean the 'Group' column by stripping any leading or trailing spaces
df1['Group'] = df1['Group'].str.strip()

# Ensure that 'Group' is treated as a string type
df1['Group'] = df1['Group'].astype(str)

# Filter df1 to only include rows where the 'Group' column ends with '0'
df1 = df1[df1['Group'].str.endswith('0')]

# Define the date columns to be formatted
date_columns = ["Ord. date", "Requested d.", "Conf. DD", "Expected DD", "DN date", "Inv. date"]

# Function to convert date columns to a specific format
def convert_date_format(df, columns):
    for column in columns:
        if column in df.columns:
            try:
                # Convert the date columns to datetime and then reformat them
                df[column] = pd.to_datetime(df[column], errors='coerce', format='%m/%d/%Y %I:%M:%S %p')
                df[column] = df[column].dt.strftime('%d/%m/%Y %I:%M %p')
            except Exception as e:
                print(f"Error processing column '{column}': {e}")
    return df

# Apply the date conversion to the dataframe
df1 = convert_date_format(df1, date_columns)

# Define the path for saving the cleaned file
output_file_path1 = "/content/cleaned_orders-overview-detail-export.xlsx"
df1.to_excel(output_file_path1, index=False)

# Print a message to confirm that the cleaned file has been saved
print(f"Cleaned file saved to {output_file_path1}")

# Load the cleaned file to continue with the next steps
df1 = pd.read_excel(output_file_path1)

# Define the path for the second file to get the column order
file2 = "/content/StarOrderplus_-_Order_items_20241230171849.xlsx"
df2 = pd.read_excel(file2)

# Get the column order from the second file (Star Orderplus)
desired_columns = df2.columns.tolist()

# Rearrange the columns in df1 to match the order from the second file
df1_rearranged = df1.reindex(columns=desired_columns, fill_value=None)

# Define the path for the reference file (Star Orderplus)
reference_file_path1 = '/content/latestStarOrderplus_-_Order_items_20240730094341 (2).xlsx'
output_file_path2 = '/content/updated_Order_items.xlsx'

# Update the column headings for the rearranged df1 and save it
update_headings(reference_file_path1, '/content/StarOrderplus_-_Order_items_20241230171849.xlsx', output_file_path2)

# Now for the second set of files (Mercedes Yedek Parça Sipariş 2024)
reference_file_path2 = '/content/latestMercedes Yedek Parça Sipariş 2024.xlsx'
target_file_path2 = '/content/Mercedes Yedek Parça Sipariş 2024.xlsx'
output_file_path3 = '/content/updated_Mercedes_Yedek_Parça_Sipariş_2024.xlsx'

# Update the column headings for the Mercedes Yedek Parça Sipariş file and save the result
update_headings(reference_file_path2, target_file_path2, output_file_path3)
```
